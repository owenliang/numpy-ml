```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 加载训练集
train = np.loadtxt('click.csv', delimiter=',', dtype='int', skiprows=1)
train_x = train[:,0]
train_y = train[:,1]
```

```python
# 特征标准化（ x - 减去均值 / 标准差 )
mean = train_x.mean()
std = train_x.std()
def standardize(x):
    return (x - mean) / std

train_x = train_z = standardize(train_x)
```

```python
# 绘制散点
plt.scatter(train_x,train_y)
```

```python
# 定义二次项回归函数 y = theta0*x1^0 + theta1*x1^1 + theta2*x1^2
theta0 = np.random.rand()
theta1 = np.random.rand()
theta2 = np.random.rand()
theta3 = np.random.rand()
def f(x):
    #return theta0 + theta1*x
    #return theta0 + theta1*x + theta2*x*x
    return theta0 + theta1*x + theta2*x*x + theta3*x*x*x
```

```python
# 训练参数
learning_rate = 0.001
# 停止条件（前后2次误差足够小）
early_stop = 0.001
# 停止条件（一定的轮数）
max_iter = 10000

# 误差函数
def E(x, y):
    return np.sum((y-f(x))**2) / 2
```

```python
# 开始训练

counter = 0
last_error = E(train_x, train_y) # 上一轮的误差
while True:
    # 随机梯度下降
    idx = np.random.randint(0, len(train_x))
    x = train_x[idx:idx+10]
    y = train_y[idx:idx+10]
    
    theta0 = theta0 - learning_rate * np.sum((f(x) - y) * 1)
    theta1 = theta1 - learning_rate * np.sum((f(x) - y) * x)
    theta2 = theta2 - learning_rate * np.sum((f(x) - y) * x * x)
    theta3 = theta3 - learning_rate * np.sum((f(x) - y) * x * x * x)
    
    cur_error = E(train_x, train_y) # 本轮训练后的误差
    print('轮数:%d 误差:%f' % (counter, cur_error))
    
    if abs(last_error - cur_error) <= early_stop: # 误差趋于稳定，退出训练
        break
    last_error = cur_error
    counter += 1
```

```python
# 绘制拟合情况
def draw_f():
    x = np.linspace(-3, 3, 100)
    y = f(x)
    plt.plot(x,y,c='green')
    plt.scatter(train_x, train_y)
draw_f()
```
